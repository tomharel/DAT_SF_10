## Functions for Logistic Regression

```python
# Prep data for logistic regression
def create_dummies(df, var_name):
    new_var_names = []
    for elem in df[var_name].unique():
        df[var_name + "_" + str(elem)] = df[var_name] == elem
        new_var_names.append( var_name + "_" + str(elem) )
    print "created {0} dummies for {1}  =>  {2}".format(len(new_var_names), var_name, new_var_names)

def add_all_titanic_dummies(df):
    # 1. Treat Pclass as dummy variables (non-parametric)
    create_dummies(df, 'Pclass')

    # 2. Embarked Location Dummies
    create_dummies(df, 'Embarked')

    # 3. Dummies for Parch and SibSp
    create_dummies(df, 'Parch')
    create_dummies(df, 'SibSp')

    df.head()

def fit_LR(model_parameters):    
    X_train = augmented_data[ model_parameters ].values
    y_train = augmented_data.Survived.values

    LR = LogisticRegression(C=regC, random_state=0).fit(X_train, y_train)

    # Results Array
    N = len(model_parameters)
    results = np.zeros((1,N+1))
    results[:,:-1] = LR.coef_
    results[:,-1] = LR.intercept_
    
    # Plot
    coef_count = np.arange(N)
    fig = plt.figure(figsize=[30,5], dpi=600)
    ax1 = fig.add_subplot(111)
    ax1.bar(coef_count, LR.coef_.ravel())
    ax1.set_xticks(coef_count + 0.5)
    ax1.set_xticklabels(model_parameters)
    timestamp = str(time.time())
    fig.savefig('coefficients' + timestamp + '.png',format='png',dpi=600)
    plt.show()
    
    parameters_and_int = model_parameters + ['intercept']
    coef_df = pd.DataFrame(results, columns=parameters_and_int)
    return coef_df, LR
```

## Cross Validation

```python
def cross_validate(X, y, classifier, k_fold, min_p = 0.5) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold(len(X), n_folds=k_fold,
                           shuffle=True, random_state=1)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ], y[ train_slice  ])
        
        
        # Classify based on threshold
        prob_vector = model.predict_proba(X[ test_slice ])[:,1]
        class_vector = []
        for i in xrange(0,len(test_slice)):
            if prob_vector[i] >= min_p:
                class_vector.append(1)
            else:
                class_vector.append(0)
           
        # count good predictions
        good_pred_count = 0
        for j in xrange(0,len(test_slice)):
            if class_vector[j] == y[ test_slice ][j]:
                good_pred_count +=1
        
        k_score = float(good_pred_count)/float(len(test_slice))
        #k_score = model.score(X[ test_slice ], y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
def cross_validate_old(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold(len(X), n_folds=k_fold,
                           shuffle=True, random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ], y[ train_slice  ])

        k_score = model.score(X[ test_slice ], y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

## ROC Curve

##### Will show a few ROC Curves so the results don't depend on the particular split.

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

def plot_roc_curve(target_test, target_predicted_proba):
    fpr, tpr, thresholds = roc_curve(target_test, target_predicted_proba[:, 1])
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate or (1 - Specifity)')
    plt.ylabel('True Positive Rate or (Sensitivity)')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="best")
    plt.savefig('roc' + str(time.time()) + '.png',format='png',dpi=600)
```
