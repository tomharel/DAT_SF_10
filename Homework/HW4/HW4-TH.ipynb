```python
from sklearn import datasets
# usual imports
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
import pandas as pd
%matplotlib inline
# scikit-learn algorithms
from sklearn.decomposition import PCA
from sklearn.decomposition import RandomizedPCA
```

```python
data = datasets.fetch_olivetti_faces()
print data.DESCR
print data
```

## PCA

```python
X, y = data.data, data.target
pca = PCA()
X_r = pca.fit(X).transform(X)
ratios = pca.explained_variance_ratio_

comp_id = [i + 1 for i in range(400)] # ids of components
fig = plt.figure(figsize=(8,5))
plt.plot(comp_id, ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.show()
```

## Randomized PCA

```python
X, y = data.data, data.target
rpca = RandomizedPCA()
X_r = rpca.fit(X).transform(X)
ratios = rpca.explained_variance_ratio_

comp_id = [i + 1 for i in range(400)] # ids of components
fig = plt.figure(figsize=(8,5))
plt.plot(comp_id, ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.show()
```

## Plot in a plane of 2 components

```python
X, y = data.data, data.target
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

plt.figure(figsize=(16,16))
plt.plot(X_r[:,0], X_r[:,1], 'g*')
```
