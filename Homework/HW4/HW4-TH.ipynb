```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pylab as pl
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.decomposition import RandomizedPCA
from sklearn import datasets

%matplotlib inline
```

```python
# Load Data
data = datasets.fetch_olivetti_faces()
```

```python
# Plot Component Variation Function
def plot_component_variation(pca):
    ratios = pca.explained_variance_ratio_
    comp_id = [i + 1 for i in range(len(ratios))] 
    fig = plt.figure(figsize=(20,8))
    plt.plot(comp_id, ratios, 'ro-', linewidth=1)
    plt.title('Scree Plot')
    plt.xlabel('Principal Component')
    plt.ylabel('Explained Variation')
    plt.xlim([0,100])
    plt.show()
```

## 1) PCA

```python
pca = PCA().fit(data.data)
plot_component_variation(pca)
```

```python
# Normalize Data
features_norm = (data.data - data.data.mean())/data.data.std()
X_norm = features_norm
pca_norm = PCA().fit(X_norm)
plot_component_variation(pca_norm)
```

```python
# I don;t see any reason to normalize. It looks pretty similar...
```

## 2) Randomized PCA

```python
rpca = RandomizedPCA().fit(data.data)
plot_component_variation(rpca)
```

```python
# Looks similar. Randomized PCA uses a random part of the data and is supposed to perform faster, but it's negligible here.
```

## 3) Plot in a plane of 2 components and color each person

```python
X, y = data.data, data.target
pca = PCA(n_components=2)
comp_plane = pca.fit(X).transform(X)

colors = matplotlib.colors.ListedColormap(np.random.rand(40,3))

plt.figure(figsize=(20,8))
for target in range(0, 40):
    plt.scatter(comp_plane[:,0][data.target == target],
                comp_plane[:,1][data.target == target],
                c=colors.colors[target])
```

```python
# I see no viable clusters here...
```
