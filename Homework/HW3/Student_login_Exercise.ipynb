```python
import numpy as np
import pandas as pd

import time
import datetime as dt
import statsmodels.api as sm
```

```python
#Read in the data
data = pd.read_csv('student_logins.csv')
```

```python
data.head(10)
```

```python
data.info()
```

```python
data['session_start_time'] = pd.to_datetime(data['session_start_time'])
data['session_end_time'] = pd.to_datetime(data['session_end_time'])
data['student_account_created'] = pd.to_datetime(data['student_account_created'])
```

```python
data.info()
```

```python
#Convert session start and end times to datetime objects
#data['session_start_time'] = data['session_start_time'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
#data['session_end_time'] = data['session_end_time'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
```

```python
#Create a timedelta object for the duration of the session
data['session_duration'] = data['session_end_time'] - data['session_start_time']
```

```python
#Convert the timedelta object to seconds
data['session_duration'] = data['session_duration'].apply(lambda x: x.item()/1000000000)
```

```python
data.head()
```

```python
#Create a new feature determining the time per problem in each session
data['time_per_problem'] = data['session_duration']/data['problems_completed']
```

```python
#Create a dictionary with the average time per problem for each class
class_tpp = {class_id: data[data['class_id'] == class_id]['time_per_problem'].mean() for class_id in np.unique(data.class_id)}
```

```python
class_tpp
```

```python
#Create a dictionary with the average time per problem for each student
student_tpp = {student_id: data[data['student_id'] == student_id]['time_per_problem'].mean() for student_id in np.unique(data.student_id)}
```

```python
student_tpp
```

```python
#Map the values from the dictionaries onto each session, 
#essentially transforming the categorical variables of class_id and student_id
#to continuous variables based on the average time per problem


data['class_avg'] = data['class_id'].map(lambda x: class_tpp[x])
data['student_avg'] = data['student_id'].map(lambda x: student_tpp[x])
```

```python
data.head(100)
```

```python
#Build a Ordinary Least Squares linear regression model based on our features

x = data[['problems_completed', 'class_avg', 'student_avg', 'student_previous_logins_total', 'student_previous_class_logins']].values
#'unique_lessons', 
y = data['session_duration'].values
X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit()
```
