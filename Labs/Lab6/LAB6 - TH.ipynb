```python
# Import Data
import pandas as pd
data = pd.read_csv("student_logins.csv")

%matplotlib inline
```

```python
# Add Duration and age at class time
from datetime import timedelta
data['session_end_time']  = pd.to_datetime(data['session_end_time'])
data['session_start_time']  = pd.to_datetime(data['session_start_time'])
data['student_account_created'] = pd.to_datetime(data['student_account_created'])
data['duration'] = data['session_end_time'] - data['session_start_time']
data['age'] = data['session_start_time'] - data['student_account_created']


# Convert Duration and Age to Seconds and Days respectively
data['duration_seconds']= data['duration'].astype("timedelta64[s]")
data['age_days']= data['age'].astype("timedelta64[D]")


# Show Data
data.head(5)
```

```python
# Plot variable of interest 
data['duration_seconds'].plot();
```

```python
# Seems like we have some outliers/bad data (seeing negative durations, and > 40k seconds ~ 1 month makes no sense)
data = data[data['duration_seconds'] > 0]
data = data[data['duration_seconds'] < 40000]

# Re-plot for verification
data['duration_seconds'].plot();
```

```python
# Looking at Covariance plots for relationship clues

from pandas.tools.plotting import scatter_matrix
scat = scatter_matrix(data, diagonal='kde', figsize=(20,20))
```

```python
# Looking at the plots, I will try to fit a linear model with all the features, and see how well each regressor performs (student_id excluded)
import statsmodels.api as sm
X = data[['problems_completed','student_previous_logins_total', 'student_previous_class_logins', 'age_days']].values
X = sm.add_constant(X, prepend=True)
y = data['duration_seconds'].values


results = sm.OLS(y, X).fit()
results.summary()
```

```python
# Test the model using mean squared error (MSE)
yhat = results.predict()
squared_error = (y - yhat)**2
MSE = squared_error.mean()

print MSE
```

```python
# Adding class category as dummies
for j in data['class_id'].unique():
    label = "class_" + str(j)
    data[label] = data['class_id'] == j

data.head(5)
```

```python
# Revised model with dummies (exlucded 'a' to avoid multicollinearity)


X = data[['problems_completed','student_previous_logins_total', 'student_previous_class_logins', 'age_days',\
          'class_e', 'class_m', 'class_g', 'class_c']].values
X = sm.add_constant(X, prepend=True)
y = data['duration_seconds'].values


results = sm.OLS(y, X).fit()
results.summary()
```

```python
# Test the model using mean squared error (MSE)
yhat = results.predict()
squared_error = (y - yhat)**2
MSE = squared_error.mean()

print MSE
```
